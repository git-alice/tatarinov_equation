```python
import sympy as sym
import numpy as np

phi1, phi2, dphi1, dphi2, t = sym.symbols('phi1, phi2, dphi1, dphi2, t',real = True)
m1, m2, l1, l2 = sym.symbols('m1, m2, l1, l2',real = True, positive = True)
a, nu, g = sym.symbols('a, nu, g',real = True, positive = True)

v_p = sym.diff(a/nu*sym.cos(nu*t),t)

v1_x = l1*dphi1*sym.sin(sym.pi/2-phi1)
v1_y = -l1*dphi1*sym.cos(sym.pi/2-phi1)

v2_x = l2*dphi2*sym.sin(sym.pi/2-phi2)
v2_y = -l2*dphi2*sym.cos(sym.pi/2-phi2)

V1_p2 = v1_x**2+(v1_y+v_p)**2
V2_p2 = (v1_x+v2_x)**2+(v1_y+v_p+v2_y)**2

T = (1/2)*(m1*V1_p2 + m2*V2_p2)
U =(m1+m2)*g*a/nu*sym.cos(nu*t)+ m1*g*l1*sym.sin(sym.pi-phi1)+m2*g*l1*sym.sin(sym.pi-phi1)+m2*g*l2*sym.sin(sym.pi-phi2)
L =T-U
```

```python
L = sym.simplify(sym.expand(L)); print(L)
```

```python
p1,p2 = sym.symbols('p1, p2')
k = sym.linsolve([p1-sym.diff(L,dphi1),p2-sym.diff(L,dphi2)],dphi1, dphi2)
n=[]*0
for j in k:
    n.append(j)
```

```python
H=p1*n[0][0]+p2*n[0][1] - L.subs([(dphi1,n[0][0]),(dphi2,n[0][1])])
H=sym.simplify(H)
H
```

```python
H_ = nu/(2*sym.pi)*sym.integrate(H,(t,0,2*sym.pi/nu))
```

```python

global i1, i2, i3, i4

i1 = sym.diff(H.subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)]),phi1)
i2 = sym.diff(H.subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)]),phi2)
i3 = sym.diff(H.subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)]),p1)
i4 = sym.diff(H.subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)]),p2)

global i1_, i2_, i3_, i4_

i1_ = sym.diff(H_,phi1).subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)])
i2_ = sym.diff(H_,phi2).subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)])
i3_ = sym.diff(H_,p1).subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)])
i4_ = sym.diff(H_,p2).subs([(m1,1),(m2,1),(l1,1),(l2,1),(g,1),(a,0.03)])
```

```python

def fH (var, time, nu_):
    return [sym.lambdify((phi1,phi2,p1,p2,t), i3.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3],time), 
            sym.lambdify((phi1,phi2,p1,p2,t), i4.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3],time), 
            sym.lambdify((phi1,phi2,p1,p2,t), i1.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3],time), 
            sym.lambdify((phi1,phi2,p1,p2,t), i2.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3],time)]

def fH_(var,t, nu_):
    return [sym.lambdify((phi1,phi2,p1,p2), i3_.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3]), 
            sym.lambdify((phi1,phi2,p1,p2), i4_.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3]), 
            sym.lambdify((phi1,phi2,p1,p2), i1_.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3]), 
            sym.lambdify((phi1,phi2,p1,p2), i2_.subs(nu,nu_), "numpy")(var[0],var[1],var[2],var[3])]

```
